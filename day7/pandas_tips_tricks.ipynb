# 16-  Splitting a string into multiple columns

```python
# importing libraries
import pandas as pd

df = pd.DataFrame({'name': ['Muhammad Raza', 'Ali Abdullah', 'Sajjad Afzal', 'Ahmed Khan'], 
                   'location': ['Lahore, Pakistan', 'Peshawar, Pakistan', 'Islamabad, Pakistan', 'Karachi, Pakistan']})
df
```

```python
# Splitting into two columns
df.name.str.split(' ', expand=True)
```

```python
# Adding those splits into new columns
df[['first_name', 'second_name']] = df.name.str.split(' ', expand=True)
df
```

```python
# Splitting the location
df.location.str.split(' ', expand=True)
```

```python
# Adding those splits
df[['city', 'country']] = df.location.str.split(' ', expand=True)
df
```

```python
# IMPORTANT!!!
# Refine data manipulation
df = df[['first_name', 'second_name', 'city', 'country']]
df
```

# 23- Reshaping a dataframe

```python
fasla = pd.DataFrame([['12345', 100, 200, 300], ['34567', 400, 500, 600], ['67890', 700, 800, 900]], columns=['zip', 'factory', 'warehouse', 'retails'])
fasla
```

```python
# We can make the dataset more longer and in detail by using melt function
fasla.melt(id_vars='zip', var_name='location_type', value_name='distance')
```

```python
sns.barplot(x='zip', y='factory', data=fasla)
```

```python
fasla_long = fasla.melt(id_vars='zip', var_name='location_type', value_name='distance')
fasla_long
```

```python
# We can not use double so that's why we use hue as location_type
# sns.barplot(x='zip', y=[['factory','warehouse','retail']], hue='location_type' ,data=fasla_long)
sns.barplot(x='zip', y='distance', hue='location_type' ,data=fasla_long)
```
