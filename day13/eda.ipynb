# Expolatory Data Analysis
Expolatory Data Analysis (EDA) is a method used to `analyze` and `summarize` datasets. In Python, you can use pandas library to perform EDA on a dataset


**Steps are the followings:**
1. Understand the data
2. Clean the data
3. Find relationship between data

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import datasets
df = sns.load_dataset('titanic')
```

```python
# Find null values
df.isnull()
```

```python
df.isnull().sum()
```

```python
# To find percentage
df.isnull().sum() * 100 / len(df)
```

```python
# PEMDAS Rule
df.isnull().sum() / len(df) * 100
```

```python
df.columns
```

```python
# To check the Normality test, Bell curve
sns.distplot(df['age'])
```

```python
# Visualize NaN values in heatmap
sns.heatmap(df.isnull())
```

```python
sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap='viridis') 
```

```python
df.drop(columns='deck', inplace=True)
```

```python
df.columns
```

```python
df.shape
```

```python
# Visualize and Explore
# Countplot for categorical values
sns.countplot(data=df, x='survived') # y='age' is not possible in countplot
```

```python
# Barplot for x = categorical y = numeric
sns.barplot(data=df, x='survived', y='fare', hue='who')
```

```python
# Data Distribution
# Histogram or Histplot for numeric values
sns.histplot(data=df, x='age', bins=30)
```

```python
df['age'].hist(bins=30, color='darkred', alpha=0.75) # alpha = opacity
```

```python
# Boxplot when = categorical, y = numeric
sns.boxplot(data=df, y='age', x='survived', hue='who')
```

```python
# Scatterplot when both numeric
sns.scatterplot(x='age', y='fare', hue='pclass', data=df) 
plt.ylim(0, 300) 
```

```python
# Pair plot used for Correlation
sns.pairplot(df, hue='survived') # To check the relationships 
```

```python
# Stacked barchart/plot
# df.groupby(['pclass', 'sex'])['survived'].mean().unstack().plot(kind='bar', stacked=True)
df.groupby(['pclass', 'who'])['fare'].mean().unstack().plot(kind='bar', stacked=True)
```

```python
# Swarm plot
sns.swarmplot(data=df, x='survived', y='age', hue='who')
```

```python
# Point plot
sns.catplot(data=df, x='pclass', y='fare', hue='embarked', kind='point')
```

```python
# Checking relation of two numeric variables
sns.lmplot(data=df, x='age', y=df[df['fare'] < 300])
# Assignment to check fare of those less than 300
sns.lmplot(data=df[df['fare'] < 300], x='age', y='fare')
# df.dtypes
```

```python
# How to add a column and make an age group
sns.histplot(data=df, x='age')
```

```python
# Grouping
df.columns
```

```python
# Create a new column
df['age_groups'] = pd.cut(df['age'], 
        [0, 10, 18, 30, 50, 80],
        labels=['Kid', 'Child', 'Youth', 'Middle Aged', 'Old'])
```

```python
df.head()
```

```python
df.columns
```

```python
df.groupby('age_groups')['fare'].mean().plot(kind='bar')
```

```python
sns.pairplot(data=df, hue='survived', vars=['age', 'fare', 'pclass'])
```

```python
# To create new column family_size based on sibsp, parch
df['family_size'] = df['sibsp'] + df['parch']
```
